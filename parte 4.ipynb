```python
import pandas as pd 

df = pd.read_csv('german_credit_data.csv')
df.head()

df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

def createXy(df, input_columns, output_column):
    
    minmax = MinMaxScaler()
    onehot = OneHotEncoder(sparse=False, drop='first') ##verificar parametros pois preguiça agora :P
    
    df.dropna(axis=0, inplace=True) ## axis=1 colunas, axis=0 linhas
    
    X = df[input_columns].copy()
    X_num = X.select_dtypes(exclude='object')
    X_num = minmax.fit_transform(X_num)
    X_num = pd.DataFrame(data=X_num, columns=X.select_dtypes(exclude='object').columns)
    
    columns = []
    X_cat = X.select_dtypes(include='object')
    X_cat = onehot.fit_transform(X_cat)
    for c, values in zip(X.select_dtypes(include='object').columns, onehot.categories_):
        for value in values[1:]:
            columns.append(f'{c}_{value}')
    print(columns)
    X_cat = pd.DataFrame(data=X_cat, columns = columns)
    
    X = X_num.join(X_cat)
    y = df[output_column].copy()
    
    return X, y

X, y = createXy(df, ['Age', 'Sex', 'Job', 'Housing', 'Saving accounts',
       'Checking account', 'Credit amount', 'Duration', 'Purpose'], 'Risk')
X
```
